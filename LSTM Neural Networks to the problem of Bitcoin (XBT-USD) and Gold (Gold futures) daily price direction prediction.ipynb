#### <font color= darkblue>Abstract:</font>



While state of the art Deep Learning algorithms based on neural networks allow an investor to predict the direction of movement of a financial asset, rather little research literature has been published on its investment applications to “alternative” currencies such as Bitcoin and Gold, which have had a phenomenal run in 2020 despite the pandemic. We consider this period a unique opportunity to study and explore the investment applications of Long Short-Term Memory (LSTM) neural networks using Keras Python package to the directional classification problem of Bitcoin (XBT/USD) and Gold (Gold futures) daily price direction prediction. We construct a neural network in Python using real data and we look at the LSTM network which involves an architectural modification to regular recurrent neural network. According to recent groundbreaking findings on Bitcoin’s microstructure market, previous research up to now on Bitcoin price discovery omits the most important price-leading instruments and information flows. Our paper presents how these pioneering findings from new published research about derivatives products traded on three large unregulated crypto exchanges that strongly dominate price discovery within the bitcoin market could enhance our LSTM model accuracy. We use a similar framework with our approach for Gold.

### <font color= darkblue> SECTION II: UNDERSTANDING LONG SHORT-TERM MEMORY (LSTM) NETWORKS</font>

#### <font color= darkblue> What is Deep Learning? </font>

Deep learning is a Machine Learning method that analyzes data in multiple layers of learning and is based on neural networks which are roughly in turn inspired by the way neurons are connected in the human brain [19]. As such, Neural Networks are structured layers of non-linear processing units – called neurons – stacked on top of each other and constitute a mathematical framework for learning representations from data [20]. In that sense, Neural Network is an architecture and Deep Learning is a mathematical framework for learning representations from data.


More specifically, the term “Deep” in “Deep Learning” stands for the idea of successive layers of neurons in the neural network and the number of layers contributing to a model of the data is called the depth of the model [21]. Once the signal weightings are calibrated from the sample dataset (training/learning dataset), these models have strong out of sample predictive power. In fact, multiple layers of signal processing units or neurons allow these models to progressively learn more complex concepts out of simpler ones. What is transformative about deep learning is that it allows a model to learn all layers of representation jointly and simultaneously rather than greedily [22]. With joint feature learning, whenever the model adjusts one of its internal features, all other features that depend on it automatically adapt to the change.


The idea of a neural network itself is not new, but the most recent developments in computer science enable us to calibrate the weights of the interconnections efficiently. Computer scientists have found that neural networks not only approximate functions well on the training set, but also generalize well to unseen examples in certain chosen applications like image text analysis (Van de Ber, 2020) [23].


These properties have made Deep Learning much more successful than previous approaches to Machine Learning. For a comprehensive and in-depth review of Deep learning, consult Y. Hilpisch (2020), F. Chollet (2017) and T. Guida (2019).

Deep Neural Networks do the input-to-target mapping via a deep sequence of simple data transformations (layers) and these data transformations are learned by exposure to examples. The specification of what a layer does to its input data is stored in the layer’s weights, also sometimes called the parameters of a layer. In others words, the transformation implemented by a layer is parametrized by its weights. Hence in this context, “learning means finding a set of values for the weights of all layers in a network, such that the network will correctly map example inputs to their associated targets” [24]. Once we initialize the weights (Initialization weights), we need to be able to measure how far this output is from what we expect and this is done by the loss function of the network (objective function) [25]. The loss function takes the predictions of the network and the true target (what we wanted the network to output) and computes a distance score, capturing how well the network has done on this specific example.

The deep learning problem then becomes an optimization problem as we are now trying to optimize the weights by minimizing the loss score yet we are also able to measure that loss. The loss function at this stage could be a SoftMax function, a cross-entropy loss function, a binary cross-entropy loss function and so on, depending on the classification problem [26]. Once we measure the Loss score, we use this score as a feedback signal to adjust the value of the weights. This adjustment is carried out by the optimizer which implements the backpropagation algorithm [27]. At this stage known as the training loop, the optimizer needs to optimize and to adjust the weights in order to minimize the loss score. Some popular optimizers (algorithms) that are used for this task are Stochastic Gradient Descent (SGD), Adam and RMSProp. Finally, a network with a minimal loss is one for which the outputs are as close as they can be to the targets: a trained network. To understand an LSTM Network, it is important to understand a Recurrent Neural Network first.


#### <font color= darkblue> Unfolding Recurrent Neural Network </font>
A Recurrent Neural Networks is understood as a generalization of a feedforward neural network except that it has a feedback loop to handle time dependencies. As such it is able to remember information in the past. Yet, a vanilla recurrent neural network is impractical to model most time series data, because it is very inefficient at retaining long-term memory. The effect of a given input on the hidden layer (and thus the output) either decays exponentially (or blows and saturates) as a function of time (or sequence length) [31]. This is due to the vanishing gradient in the back-propagation through time algorithm, an algorithm used to estimate the weights in the neural networks using gradient descent, as mentioned earlier. As gradients propagated back in time tend to vanish, the weights do not get updated and past information becomes irrelevant (o even worse, in case the gradients do not vanish, they tend to blow up and the estimated weights will oscillate and diverge) [32].


In fact, better designs of recurrent neural networks that attempt to solve the problem above consist of various gates to control the flow of data across the network. Hence the need for an alternative approach; the most popular architectures include the Long Short-Term Memory (LSTM) neural network [33], Gated Recurrent Unit (GRU) [34] and the more recent Generative Adversarial Networks (GAN). For more background information on recurrent neural networks, refer to the books by Hilpisch (2020), Goodfellow et al. (2016) and Chollet (2017), respectively.

#### <font color= darkblue> Long Short-Term Memory Network & Time Series Analysis:</font>
Econometricians rely on “Autoregressive integrated moving average” (ARIMA) models for forecasting time-series while within Deep learning, the analogue to ARIMA is Long Short-Term Memory. Hochreiter and Scmidhuber (1997) originally introduced the LSTM network, similar to the Recurrent Neural Network but where the hidden units are replaced by memory cells. In fact, the LSTM network is a set of subnets with recurrent connections, known as memory blocks. The hidden cell in the LSTM is a structure that holds an internal state with a recurrent connection of constant weigh which allows the gradients to pass many times without exploding or vanishing [35]. After describing the basic architecture of an LSTM network below; we provide an example of a possible use of LSTM in forecasting asset prices direction i.e., Bitcoin and Gold in subsequent sections.

#### <font color= darkblue> Mathematical Formulation of LSTM block</font>
The basic module in a LSTM network is different from a conventional neuron. Architecture of a single LSTM neuron is illustrated in the figure below: 


#### Figure 2: Flowchart taken from Singaravelu K. (2020) Introduction to Deep Learning & Neural Networks, Lecture slide 84. 

<img src="image2-Copy1.png">

LSTMs have four interacting layers and use gates to regulate the flow of information. Here, $h_t$ represents the output of the neuron and $C_t$ represents the state of the neuron. The operation can be explained as follows:

• Calculate what needs to be forgotten at the cell state:

\begin{align}f_{t} = \sigma (W_f [h_{t-1}, x_t] + b_f)\end{align}. 

• Calculate what need to be added to cell state via 

\begin{align}i_{t} = \sigma (W_i [h_{t-1}, x_t] + b_i)\end{align}and \begin{align}\overline{c}_t = tanh (W_c [h_{t-1}, x_t] + b_c)\end{align}.

• Update the cell state as (cell gate)	

\begin{align} C_t = f_t*C_{t-1} + i_t *\overline{c}_{t-1} \end{align}.

• Compute the output via 

\begin{align}o_t = \sigma (W_o [h_{t-1}, x_t] + b_o)\end{align} and \begin{align}h_t = o_t * tanh (C_t)\end{align}

As mentioned above, the structure of a LSTM neural network is one which is composed of a memory unit $Ct$, a hidden state $ht$ and three types of gates, where t indexes the time step. Specifically, for each step $t$, LSTM receives an input $xt$ and the previous hidden state $h_{t-1}$ then calculates the activation of the gates. Finally, the memory unit and the hidden state are updated.

#### <font color= darkblue> Bitcoin Market Structure and Price Discovery </font>

#### <font color= darkblue> Bitcoin Market Structure </font>


The spectrum of regulated bitcoin derivatives has been extended by the physically-settled futures traded on Bakkt, a subsidiary of the Intercontinental Exchange (ICE), and most recently also with bitcoin options offered by both (Chicago Mercantile Exchange) CME and Bakkt. Besides regulated bitcoin derivatives there are numerous online centralized crypto exchanges that allow traders to operate in totally unregulated markets. The venues with the greatest trading volumes (including BitFinance, Huobi, OKEx and Deribit) all offer similar derivatives products, including futures, options and perpetual contracts, mostly on bitcoin but also on several other crypto-asset underlyings.


Bitcoin and other cryptographic coins and tokens form a new class of financial asset with some very novel and interesting properties, including the market microstructure. They can be traded 24/7 and 365 days per year, either ‘off-chain’ on hundreds of different centralized exchanges or transferred directly ‘on chain’ using smart contracts on the Ethereum chain or other public blockchains via peer-to-peer (P2P) trading venues called decentralized exchanges [42]. Few of these venues are regulated, yet a wealth of limit order and transaction data are freely available, providing a level of transparency that is quite unique. Moreover, most platforms have an API which allows order book data to be downloaded directly at very high frequency. Data and software provider CoinAPI links with hundreds of crypto spot and derivatives exchanges, offering historical and streaming order book and trades in tick-by-tick or highest granularity data from all major centralized (and decentralized) exchanges. 

As a result, crypto assets and their derivatives have become the ideal testing ground for studying market microstructure and price discovery with bid-ask spreads, inter-exchange spreads and relative trading volumes becoming important determinants of price discovery.

#### <font color= darkblue> Price Discovery: Bitcoin Spot and Derivatives Markets, which market leads the other in pricing?</font>

The industry has matured ultimately leading to the predominant use of derivatives to express a view on the market. Among crypto derivatives, Bitcoin futures went first mainstream and are among the most popular traded instruments in the space. Further evidencing the maturation of the product, options on the futures were added. A leading futures market suggests the existence of a robust base of traders who may trade on such market for many reasons such as trust in the exchange venue and lower latency. Up until recently the infinite depth order book data in financial markets was generally hard to obtain. With the arrival of cryptocurrency exchanges with open APIs it became possible to capture all the incoming orders and reconstruct the complete order book, inciting new insight on the investor behavior and liquidity dynamics in cryptocurrency market [43].


As result, Bitcoin futures have grown to become a significant part of the bitcoin market, demonstrated not only through trading volume and open interest but also by influence on bitcoin price formation and discovery. However, price discovery in Bitcoin markets remains a contested topic. Academic research on the microstructure of bitcoin markets has become prolific and price discovery remains a fundamental research question. [44], [45] and [46] provide an overview upon the role of futures in bitcoin price discovery.


When trying to determine whether price discovery occurs in bitcoin’s spot market or bitcoin’s derivatives market, specifically focusing on futures and perpetual swaps, researchers have concluded that derivatives market more frequently leads price discovery of Bitcoin. Academics analyzing this lead/lag relationship between bitcoin spot and futures markets have found that the majority of price discovery takes place in the futures market. Research indicated that the futures market often leads the spot market due to inherent leverage, low transaction costs, the absence of short-selling restrictions and greater transparency [47]. Further analyses showed the spot market is more likely to indicate the direction of price movement while the derivatives market indicates the magnitude of the price movement [48].

In her presentation “Bitcoin Derivatives: the impact of Unregulated Markets on Price Discovery and Introducing the BVIX” at the 6th Quant Insights Conference, Carol Alexander presented results from the paper “Price Discovery in Bitcoin: The Impact of Unregulated Markets” (C. Alexander and D. Heck, 2020). Their approach is different from previous research in the sense that the authors conduct their analyses in multiple dimensional information flows while other published studies on price discovery in Bitcoin markets are limited to the spot market or rely on just two-dimensional models including one regulated future and one spot exchange. What is interesting about this research approach is that it is the first study to analyze information flows between unregulated bitcoin derivatives and the CME and Bakkt futures, as well as the major spot exchanges. They examine 21 different bitcoin spot and derivatives markets with a main focus on the price discovery role played by the heavily traded derivatives products offered by a large number of unregulated crypto exchanges. It is a thoroughly investigation of crypto exchange microstructure to develop a model for the time-lag between large price moves on leading derivatives exchanges such as Huobi, OKeX and BitMex and Bitcoin spot exchanges and other Bitcoin derivatives exchanges (e.g. CME futures). These groundbreaking results show that Bitcoin derivatives (perpetual swaps and futures) traded on the unregulated exchanges Huobi, OKEx and BitMex are by much the strongest instruments for bitcoin price discovery. Therefore, 4 unregulated bitcoin derivatives products such as the Huobi futures, OKEx futures, BitMEX perpetual and OKEx perpetual are the four leading instruments, as those products combined account for more than 60\% of total price discovery in the Bitcoin market.

#### <font color= darkblue>Volatility Indicator - Average True Range (ATR) </font>
Average True Range (ATR) is a volatility  indicator. The average true range is an $n$-period average of the true range values. An expanding ATR indicates increased volatility or range of the underlying, while a low ATR value indicates a series of periods with small ranges (quiet days). ATR based exit or stops are popular as it signals changes in volatility.

True range is defined as

$TR = max[(high-low), \space abs(high-close_{prev}), \space abs(low-close_{prev})]$


ATR is defined as


$ATR = \frac{1}{n}\sum_{i=1}^n TR_i$



where $n$ is the number of period

```python
# Function to calculate average true range (ATR) 
# We switch 'Close' FOR 'Open' because our DATAFRAME does not have data on 'Close'
def ATR(df,n):
    "function to calculate True Range and Average True Range"
    df = df.copy() 

    df['H-L']  = abs(df['High']-df['Low'])
    df['H-PO'] = abs(df['High']-df['Open'].shift(1))
    df['L-PO'] = abs(df['Low']-df['Open'].shift(1))
    
    df['TR']   = df[['H-L','H-PO','L-PO']].max(axis=1,skipna=False)
    df['ATR']  = df['TR'].rolling(n).mean()

    df2 = df.drop(['H-L','H-PO','L-PO'],axis=1)
    return df2
```

```python
# we adjust the dimensions
Bitcoin=Bitcoin.rename(columns={'Last':'Open'}) 
ATR_Bitcoin = ATR(Bitcoin,21)['ATR']
Bitcoin=Bitcoin.rename(columns={'Open':'Last'}) 
Bitcoin['Returns']=daily_returns_Bitcoin
Bitcoin['ATR']=ATR_Bitcoin
```

#### <font color= darkblue>On Balance Volume (OBV) </font>

```python
#OBV
def OBV(df):
    n=len(df)
    OBV=[]
    OBV.append(0)
    for i in range(0,n-1):
        if df['Price'][i]<df['Price'][i+1]:
            OBV.append(OBV[i]+df['Volume'][i+1])
        elif df['Price'][i]>df['Price'][i+1]:
            OBV.append(OBV[i]-df['Volume'][i+1])
    return OBV
```

#### <font color= darkblue> $K$- Means Clustering </font>
Clustering is a branch of unsupervised machine learning models that seeks to learn from the properties of the data by identifying groups or clusters in the dataset. Clustering is different from classification as it seeks to identify a set of similar data points and calling the resulting set a cluster. 

##### <font color= darkblue> $k$-Means Algorithm </font>                                                                                                                           
The $K$-means algorithm searches for a predetermined number of clusters within an unlabeled dataset and is based on the assumptions that the optimal cluster will have:

- “cluster center” which is the mean of all the points belonging to the cluster.

- each point is closer to its own cluster center than to other cluster centers.


The "E-Step" or "Expectation Step" assigns each data point to its nearest centroid. Then, the "M-Step" or Maximization Step" computes the mean of all the points for each cluster and sets the new centroid. The $k$-means algorithms seeks to find $k$ clusters within a data set. The clusters are chosen to reduce the inertia, the objective function

$\min_{C_k} \sum_{k} \sum_{X_j \in  C_k} \| X_j  - \mu_k \|^2.$

and the centroid of a cluster $\mu_k$ is equal to

$\mu_k = \frac{1}{|C_k|} \sum_{X_j \in C_k} X_j,$

```python
kmeans = KMeans(
    init="random",
   n_clusters=len(Bitcoin.keys()),
   n_init=len(Bitcoin.keys()),
   max_iter=300,
   random_state=42)
kmeans.fit(Bitcoin)

#Choosing the Appropriate Number of Clusters
kmeans_kwargs = {
    "init": "random",
    "n_init": len(Bitcoin.keys()),
    "max_iter": 300,
    "random_state": 42,
    }
# A list holds the SSE values for each k
sse = []
for k in range(1, len(Bitcoin.keys())):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(Bitcoin)
    sse.append(kmeans.inertia_)

plt.style.use("fivethirtyeight")
plt.plot(range(1, len(Bitcoin.keys())), sse)
plt.xticks(range(1, len(Bitcoin.keys())))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

#### <font color= darkblue> Self Organizing Maps (SOM)</font>
Self-organizing maps (SOM) are a class of unsupervised learning neural networks used for feature detection. SOM are used to project a high dimensional data into a smaller set and are used for dimensionality reduction.

SOMs differ from conventional artificial neural networks as they apply competitive learning when compared to error correlated learning, which involves backpropagation and gradient descent. In SOM, the training data has no labels and the map learns to differentiate and distinguish features based on similarities.

##### <font color= darkblue>  SOM Algorithm </font>
Input : $X_i, 1 \leq i \leq n$

Output : $W, C_j, 1 \leq j \leq k$

Step 1 : Initialize weights $w_{ij}$

Step 2 : Calculate distance $d_{ij} = \sum (w_{ij} - x_i)^2$

Step 3 : Calculate $w_{ij}(t+1) = w_{ij}(t)(1 - \alpha) + \alpha x_i$

where $n$ is number of features, $k$ is number of clusters and $t$, the iteration.

```python
# Initialize a 10x10 SOM
size = len(list(Bitcoin.keys()))
som = MiniSom(size, size, len(list(Bitcoin.keys())), learning_rate=0.5, sigma=5, random_seed=42) 
# Transfor feature matrix
scaler = MinMaxScaler()
scaled_Bitcoin = scaler.fit_transform(Bitcoin)
# Initialize with randon weights 
som.random_weights_init(scaled_Bitcoin) 
# Trains SOM with 10000 iterations 
som.train_batch(scaled_Bitcoin,10000, verbose=True) 

# Plot SOM 
plt.figure(figsize=(20, 10)) 
for ix in range(len(scaled_Bitcoin)): 
	winner = som.winner(scaled_Bitcoin[ix]) 
	#plt.text(winner[0], winner[1],names[ix], bbox=dict(facecolor='white', alpha=0.5, lw=0))  
plt.imshow(som.distance_map()) 
plt.grid(False) 
plt.title('Self Organizing Maps') 
plt.show()
```

#### <font color= darkblue> Define Model
A neural network is made up of two parts. The choice of the number of neurons, hidden layers, and activation functions determines how $x$ variables are transformed into $y$, given some fixed weights. The transformation process is known as the
forward propagation step.

First, we define the LSTM Recurrent Neural Network by initializing the Sequential Model. We'll add more LSTM layers and adjust the Dropout in order to improve the accuracy of our model. Finally, a Dense layer is used as output layer.

Next, we compile and train the model defined below. Iteratively, we can increase or decrease the epochs and batch size to get more accuracy.

#### <font color= darkblue> LSTM Model Creation </font>

Similar to Sklearn, the core data structure used in Keras is a model. The simplest type of model is the Sequential model which is a simple linear stack of layers.




We define a model\_() function to build a LSTM model in Keras with a specified number of hidden layers (hidden\_layers), each with a specified number of neurons (neurons) and a dropout percentage (drop_pct).The function returns the compiled model.The edge of the code presented below is the possibility of easily implement any change that need to be done without re-writing entire bits of the code script. 



```python
def model_(X,optimizer,activation,loss,metrics):
    #optimizer=str(optimizer)
    #activation=str(activation)
    #loss=str(loss)
    # Define the Model

    # Instantiate the Sequential Model
    model = Sequential()

    # Add first layer with dropout regularisation with 100 neurons and inputs shape for 1st set
    model.add(LSTM(units=256, input_shape = (X.shape[1],X.shape[2]), return_sequences=True)) 
    model.add(Dropout(0.4))

    # Add second layer with dropout
    model.add(LSTM(units=256, return_sequences=False))
    model.add(Dropout(0.4))

    # Add a Dense layer
    model.add(Dense(64,  activation = activation))
    # activation = 'softmax' has a lower error

    # Add the output layer
    model.add(Dense(1))


    # Compiling the model
    model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

    # Fitting the Model
    # Training the model to the training dataset
    #model.fit(X_train, y_train, batch_size=64, epochs=200, verbose=1, validation_split=0.05, shuffle=False)
    return model
```

```python
# Define the Model
metrics=[keras.metrics.Accuracy(),
    keras.metrics.BinaryAccuracy(name="binary_accuracy", dtype=None, threshold=0.5),
    keras.metrics.MeanSquaredError(name='my_mse'),
    keras.metrics.BinaryCrossentropy(),
    keras.metrics.Hinge()]
```

#### <font color= darkblue> Fitting the Model</font>

The classic Stochastic gradient descent (SGD) method has been enhanced by various adaptive moment estimation (ADAM) techniques that build on AdaGrad and RMSProp \& others. There is also some *Practical Advice for Building Deep Neural Networks* by Matt H and Daniel R that explains why the ADAM optimiser, ReLU activation function and variance scaled initialisation are preferred.
 
 In fact, while Stochastic Gradient Descent is used in many Neural Network problems, it has the problem of converging to a local minimum. This of course presents a problem considering Bitcoin price. Some other nice optimizers are variations of adaptive learning algorithms, like Adam, Adagrad, and RMSProp. Adam was found to work slightly better than the rest.

```python
# # Training the model to the training dataset
model=model_(Bitcoin_train, 'adam','relu','mean_squared_error',metrics) 
history=model.fit(Bitcoin_train, y_train, batch_size=64, epochs=200, verbose=1, validation_split=0.05, shuffle=False)
```

Training on small sample sizes through so-called generative-adversarial models (GAN) is still at an incipient research stage. 


Neural network models and deep learning algorithms are very fragile in the sense that noise into input set can affect predictions’ validity. This is due to a gap in the way the neural network trains the algorithms. This issue can be fixed by producing a model that generates “fake data” and then parsing the “fake data” and comparing that data with the original data set while using the model prediction output. Hence, Generative Adversarial Networks and fake data; as a way to generate a new data set (fake data set) could offer improved accuracy to the model.
